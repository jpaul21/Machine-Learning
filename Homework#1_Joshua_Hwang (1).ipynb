```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples,1) - 0.8)
y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples,1)
plt.plot(x,y,'ro')
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
degree = 6
x = 25*(np.random.rand(number_of_samples,1)-0.8)
y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples,1)
p = np.polyfit(x.flatten(),y.flatten(),degree)
z = np.poly1d(p)
plt.plot(x,y,'ro')
plt.plot(np.sort(x,axis=None),z(np.sort(x,axis=None)))
plt.show()

```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
degrees = np.arange(1,9,1)
x = 25*(np.random.rand(number_of_samples,1)-0.8)
y = 7*x + 15*x**2 + x**3 + noise_scale*np.random.randn(number_of_samples,1)
errors = []
for a in range(8):
    p = np.polyfit(x.flatten(),y.flatten(),a)
    z = np.poly1d(p)
    mse = np.square(np.subtract(y,z(x))).mean()
    errors.append(mse)
plt.plot(degrees,errors)
plt.show()
```
